```python
# Install the sentence-transformers library
!pip install -q sentence-transformers

# Import the necessary classes
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
# Load a pre-trained model
# This model is great for its balance of speed and performance.
model = SentenceTransformer('all-MiniLM-L6-v2')

print("Model loaded successfully!")
```

```python
# This is our "database" of sentences we want to search.
corpus = [
    'A man is eating a piece of bread.',
    'The girl is petting a fluffy cat.',
    'A cheetah is running at high speed across the savannah.',
    'The new AI technology from Google is revolutionizing the industry.',
    'Someone is strumming a guitar.',
    'Paris is famous for the Eiffel Tower.',
    'How can I learn to play the piano?'
]
```

```python
# Encode the corpus sentences into embeddings
# This creates a vector representation for each sentence.
corpus_embeddings = model.encode(corpus, convert_to_tensor=True)

# Let's check the shape of our embeddings
print("Shape of corpus embeddings:", corpus_embeddings.shape)
# The output should be [7, 384], meaning 7 sentences, each with a 384-dimensional vector.
```

```python
# Our search query
query = "What is france reknowned for?which city"

# 1. Encode the query into an embedding
query_embedding = model.encode(query, convert_to_tensor=True)

# 2. Compute cosine similarity between the query and all corpus sentences
# This function is highly optimized to compare the query vector against all corpus vectors.
cos_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]

# 3. Find the top 3 most similar sentences
# We use torch.topk to find the highest scores and their indices.
top_results = torch.topk(cos_scores, k=3)

print(f"Query: \"{query}\"\n")
print("Top 3 most similar sentences in the corpus:")

# 4. Print the results
for score, idx in zip(top_results[0], top_results[1]):
    # The score is the cosine similarity. The idx is the position in our original corpus.
    print(f"- {corpus[idx]} (Score: {score:.4f})")
```

```python
# Another example to show understanding of concepts
query_2 = "A person is making music"

query_embedding_2 = model.encode(query_2, convert_to_tensor=True)
cos_scores_2 = util.cos_sim(query_embedding_2, corpus_embeddings)[0]
top_results_2 = torch.topk(cos_scores_2, k=3)

print(f"\nQuery: \"{query_2}\"\n")
print("Top 3 most similar sentences in the corpus:")

for score, idx in zip(top_results_2[0], top_results_2[1]):
    print(f"- {corpus[idx]} (Score: {score:.4f})")
```
